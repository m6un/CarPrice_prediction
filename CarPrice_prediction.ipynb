```python
#importing the neccesarry libs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(threshold = np.inf)
pd.set_option('display.max_columns', None)
```

```python
#reading the csv 
df = pd.read_csv('modified_car_price_variables1.csv')
df = df.drop(df.columns[0], axis=1)
df
```

```python
#label_encoding
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df[['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'doornumber', 'cylindernumber']] = df[['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'fuelsystem', 'doornumber', 'cylindernumber']].apply(le.fit_transform)
df
```

```python
#defining X and y
X = df.iloc[:, 1:-1]
y = df.iloc[:,-1]
```

```python
#train test splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
X_train, y_train
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
X_train = mm.fit_transform(X_train)
X_test = mm.transform(X_test)
X_train
```

# Random Forest

```python
#importing RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
#initialising and fitting
regressor = RandomForestRegressor(n_estimators=50, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score:', metrics.r2_score(y_test, y_pred)) 
```

### Visualisation

```python
#Visualizing single decision tree
#import tools needed for visualisation
rf = regressor
features = df.drop(['price', 'CarName'], axis = 1)

# Saving feature names for later use
feature_list = list(features.columns)
# Convert to numpy array
features = np.array(features)
feature_list

```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot

# Limit depth of tree to 3 levels
rf_small = RandomForestRegressor(n_estimators=10, max_depth = 3)
rf_small.fit(X_train, y_train)
# Extract the small tree
tree_small = rf_small.estimators_[5]
# Save the tree as a png image
export_graphviz(tree_small, out_file = 'small_tree.dot',
                feature_names = feature_list, rounded = True, precision = 1, proportion = False, filled= True)

(graph, ) = pydot.graph_from_dot_file('small_tree.dot')
graph.write_png('small_tree.png');
```

```python
importances = list(rf.feature_importances_)


# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 3)) for feature, importance in zip(feature_list, importances)]
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

# Linear Regression and Ridge Regression

```python
#importing libraries 
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error as mse
lr = LinearRegression()
lr.fit(X_train, y_train)
#predicting values
y_pred_lr = lr.predict(X_test)
rmse_lr = np.sqrt(mse(y_test, y_pred_lr))
rr = Ridge()
rr.fit(X_train, y_train)
#predicting values
y_pred_rr = rr.predict(X_test)
rmse_rr = np.sqrt(mse(y_test, y_pred_rr))
```

```python
#Evaluating the predictions
print('-------------Linear Regression-------------')
print("RMSE = {:.2f}".format((rmse_lr)))
accuracy1 = lr.score(X_test,y_test)
print('Accuracy = ', round(accuracy1*100, 2),'%')
print('-------------Ridge Regression--------------')
print("RMSE = {:.2f}".format((rmse_rr)))
accuracy2 = rr.score(X_test, y_test)
print('Accuracy = ', round(accuracy2*100, 2),'%')
```
